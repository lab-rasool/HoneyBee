## Pathology Reports

```python
from honeybee.loaders import PDFreport
from honeybee.models import HuggingFaceEmbedder

DATA_DIR = "/mnt/d/TCGA/raw/TCGA-ACC"
MANIFEST_PATH = DATA_DIR + "/manifest.json"
MODALITY = "Pathology Report"
PARQUET = f"/mnt/d/TCGA/parquet/{MODALITY}_gatortron-base.parquet"

pdf_report = PDFreport(chunk_size=512, chunk_overlap=10)
```

```python
from honeybee.loaders import manifest_to_df

df = manifest_to_df(MANIFEST_PATH, MODALITY)
df
```

```python
from rich.console import Console
from rich.markdown import Markdown
from rich.panel import Panel

console = Console()

embedding_model = HuggingFaceEmbedder(
    model_name="UFNLP/gatortron-base",
    pooling_method=None,
    max_length=512,
)


for idx, row in df.iterrows():
    PatientID = row["PatientID"]
    id = row["id"]
    filename = row["file_name"]
    path = f"{DATA_DIR}/raw/{PatientID}/{MODALITY}/{id}/{filename}"

    text = pdf_report.read(path)
    embeddings = embedding_model.generate_embeddings(text)

    console.print(Panel(Markdown(text), title=f"{PatientID}/{id}"))
    console.print(embeddings.shape)

    if idx > 5:
        break
```

## Whole Slide Images

```python
import gc
import numpy as np
import pyarrow as pa
import pyarrow.parquet as pq
import torch
from tqdm import tqdm

from honeybee.loaders import Slide, manifest_to_df
from honeybee.models import UNI, TissueDetector

DATA_DIR = "/mnt/d/TCGA/raw/TCGA-ACC"
MANIFEST_PATH = DATA_DIR + "/manifest.json"
MODALITY = "Slide Image"
PARQUET = f"{MODALITY} (UNI).parquet"

HE_DETECTOR_PATH = "/mnt/d/Models/TissueDetector/HnE.pt"
EMBEDDING_MODEL_PATH = "/mnt/d/Models/UNI/pytorch_model.bin"
df = manifest_to_df(MANIFEST_PATH, MODALITY)

tissue_detector = TissueDetector(model_path=HE_DETECTOR_PATH)
embedding_model_path = EMBEDDING_MODEL_PATH
uni = UNI()

df["embedding"] = None
df["embedding_shape"] = None
writer = None
for index, row in tqdm(df.iterrows(), total=len(df), desc="Processing"):
    try:
        slide_image_path = f"{DATA_DIR}/raw/{row['PatientID']}/{MODALITY}/{row['id']}/{row['file_name']}"
        slide = Slide(
            slide_image_path,
            tileSize=512,
            max_patches=100,
            visualize=False,
            tissue_detector=tissue_detector,
        )
        patches = slide.load_patches_concurrently(target_patch_size=224)

        if patches.shape[0] == 0:
            slide = Slide(
                slide_image_path,
                tileSize=512,
                max_patches=1000,
                visualize=True,
                tissue_detector=tissue_detector,
            )
            patches = slide.load_patches_concurrently(target_patch_size=224)

            if patches.shape[0] == 0:
                with open("errors.txt", "a") as f:
                    f.write(f"{slide_image_path} | No patches extracted.\n")
                raise ValueError("No patches extracted.")

        embedding = uni.load_model_and_predict(embedding_model_path, patches)
        df.at[index, "embedding_shape"] = embedding.shape
        embedding = embedding.reshape(-1)
        embedding = np.array(embedding, dtype=np.float32)
        embedding = embedding.tobytes()
        df.at[index, "embedding"] = embedding
    except Exception as e:
        with open("errors.txt", "a") as f:
            f.write(f"{slide_image_path} | {e}\n")
        df.at[index, "embedding"] = None
        continue

    if writer is None:
        table = pa.Table.from_pandas(df.iloc[[index]])
        writer = pq.ParquetWriter(PARQUET, table.schema)
    else:
        table = pa.Table.from_pandas(df.iloc[[index]])
        writer.write_table(table)

    del slide, patches, embedding, table
    gc.collect()
    torch.cuda.empty_cache()

if writer is not None:
    writer.close()
```
